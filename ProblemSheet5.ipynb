<b>Problem 1.</b> Let us consider the following twp point BVP

\begin{cases}
-u''(x)=1,\quad x\in(0,\pi/4)\\
u(0)=1\\
u(\pi/4)=2
\end{cases}

(a) Apply the standard second order finite difference method approximating the second order derivate. Make a plot about the numerical and the analytical solutions.
<br>

```octave
a=0; b=pi/4; %interval
alpha=1; beta=2; %boundary values
N=16;
h=(b-a)/(N+1);
e=ones(N+2,1);
A_h=(1/h^2)*spdiags([-e 2*e -e],[-1,0,1],N+2,N+2);
spy(A_h)
```

```octave
a=0; b=pi/4; %interval
alpha=1; beta=2; %boundary values
% Uniform grid
N=32;
h=(b-a)/(N+1);
% A_h
e=ones(N+2,1);
A_h=(1/h^2)*spdiags([-e 2*e -e],[-1,0,1],N+2,N+2);
A_h(1,1)=1; A_h(1,2)=0;
A_h(N+2,N+1)=0; A_h(N+2,N+2)=1;
%f right-hand side
f=ones(N+2,1); f(1)=alpha; f(N+2)=beta;
% Numerical and exact
y=A_h\f;
x=linspace(a,b,N+2)';
exact=-0.5*x.^2+(4/pi+pi/8)*x+1;
plot(x,y,'r+',x,exact)
legend('Numerical','Exact','Location','southeast')
%maxnorma=norm(y-exact,'inf')
```
